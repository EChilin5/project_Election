```
import requests
from bs4 import BeautifulSoup
import csv
import os.path
from os import path
import pandas as pd

def readStateData(state):

    url = "https://projects.fivethirtyeight.com/polls/president-general/"+state+"/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    container_1 = soup.find_all(class_='date-wrapper')
    container_4 = soup.find_all(class_='heat-map')

    biden_votes = []
    trump_votes = []
    dates = []
    data = []

    for items in container_1:
        dates.append(items.get_text())

    for items in container_4:
        data.append(items.get_text())

    update_data =[]
    for votes in range (len(data)):
        num =votes
        if (num % 2) == 0:
            update_data.append(data[votes])
   

    for votes in range (len(update_data)):
        num =votes
        if (num % 2) == 0:
            biden_votes.append(update_data[votes])
        elif (num % 2) != 0:
            trump_votes.append(update_data[votes])

    if(path.exists('president_output.csv') == False):
        with open('president_output.csv', mode = 'w') as file_csv:
            field_names =['State', 'Biden_Supporters', 'Trump_Supporters']
            writer = csv.DictWriter(file_csv, fieldnames=field_names)
            writer.writeheader()
            for i in range(len(biden_votes)-1):
                writer.writerow({'State': state, 'Biden_Supporters': str(biden_votes[i]), 
                'Trump_Supporters': str(trump_votes[i])})
    else:
        with open('president_output.csv', 'a') as new_file_csv:
            add_writer = csv.writer(new_file_csv)
            for i in range(len(biden_votes)-1):
                add_writer.writerow([state, str(biden_votes[i]), str(trump_votes[i])])

            



states = ["alabama", "alaska", "arizona", "arkansas", "california", "colorado", "connecticut",\
          "delaware", "florida", "georgia", "hawaii", "idaho", "illinois", "indiana", "iowa", \
          "kansas", "kentucky", "louisiana", "maine", "maryland", "massachusetts", "michigan",\
          "minnesota", "mississippi", "missouri", "montana", "nebraska", "nevada", "new-hampshire",\
           "new-jersey", "new-mexico", "new-york", "north-carolina", "north-dakota", "ohio", \
           "oklahoma", "oregon", "pennsylvania", "rhode-island", "south-carolina", "south-dakota", \
           "tennessee", "texas", "utah", "vermont", "virginia", "washington", "west-virginia", \
           "wisconsin", "wyoming", "district-of-columbia"]

for state in states:
	state_data = (readStateData(state))
```
